```python
import numpy as np
import pandas as pd
h = 6.654e-34
e = 1.60217657e-19
Vf = 8e5
N = 3
Lg_range = np.linspace(1e-6, 1e-8, num=10000)


t = np.linspace(500, 5000, 10000)
b = np.linspace(500, 5000, 10000)
l = np.linspace(20, 100, 10000)
r = np.linspace(20, 100, 10000)


Ncht = t * 0.6 * e / e
Nchb = b * 0.6 * e / e
Nchl = l * 0.6 * e / e
Nchr = r * 0.6 * e / e


data = []

for Lg, Ncht_val, Nchb_val, Nchl_val, Nchr_val in zip(Lg_range, Ncht, Nchb, Nchl, Nchr):
    Lg_calculated = ((h / (4 * e ** 2 * Vf)) * (1 / Ncht_val + 1 / Nchb_val + 1 / Nchl_val + 1 / Nchr_val)) * 1e-3

    data.append([Ncht_val, Nchb_val, Nchl_val, Nchr_val, Lg_calculated])

np.savetxt('graphene_inductance2.csv', data, delimiter=',', header='Ncht,Nchb,Nchl,Nchr,Lg', comments='')
```
