```python
import numpy as np


h = 6.654e-34  # Planck constant (JÂ·s)
Vf = 8e5  # Fermi velocity (m/s)
Pgr = 1.5
Ef = 0.6 * 1.60217657e-19  # Fermi energy (J)
e = 1.60217657e-19  # Elementary charge (C)
N = 3  # Number of branches
t = np.linspace(500, 5000, 10000)
b = np.linspace(500, 5000, 10000)
l = np.linspace(20, 100, 10000)
r = np.linspace(20, 100, 10000)


Ncht = t * Ef / e
Nchb = b * Ef / e
Nchl = l * Ef / e
Nchr = r * Ef / e


Wgr_range = np.linspace(500e-9, 1e-9, num=10000)
Ledge_data = []
for Wgr in Wgr_range:
    Ledge = 0
    for m in range(1, 3):
        Ledge += (Wgr / Pgr) * np.sqrt(((2 * Wgr * Ef / (m * h * Vf))**2) - 1)
    Ledge_data.append(Ledge)

# Constants for Lac
Vs = 20e3  # Source velocity (m/s)
Kb = 1.38e-23  # Boltzmann constant (J/K)
T = 300  # Temperature in Kelvin
Dac = 18 * e  # Acceptor doping concentration (C/m^3)
Ns = 5e16  # Electron concentration (m^-3)

# Calculate Lac
Lac = (1 / (Kb * T)) * (Vs**2 / Dac**2) * np.sqrt(np.pi / Ns)

# Calculate Lamdaeff
Lamdaeff_data = 1 / ((1 / Lac) + (1 / np.array(Ledge_data)))

# Calculate resistance components
Rgrt_data = (h / (2 * e**2 * Ncht)) * ((1 / Lamdaeff_data))
Rgrb_data = (h / (2 * e**2 * Nchb)) * ((1 / Lamdaeff_data))
Rgrl_data = (h / (2 * e**2 * Nchl)) * ((1 / Lamdaeff_data))
Rgrr_data = (h / (2 * e**2 * Nchr)) * ((1 / Lamdaeff_data))

# Calculate total resistance
Rgr_data = 1 / ((1 / Rgrt_data) + (1 / Rgrb_data) + (1 / Rgrl_data) + (1 / Rgrr_data))

# Save data to CSV
graphene_resistance_data = np.column_stack((Wgr_range, Rgrt_data, Rgrb_data, Rgrl_data, Rgrr_data, Rgr_data))
header = 'Wgr , Rgrt , Rgrb , Rgrl , Rgrr , Rgr'
np.savetxt('graphene_resistance.csv', graphene_resistance_data, delimiter=',', header=header, comments='')
```
