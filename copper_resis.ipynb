```python
import numpy as np

Rc = 57.7e+04
wc = 22.0e-09
tc = 44.0e-09

rowcu = Rc * wc * tc


Wcu = np.linspace(50, 140, 100)
Tcu = np.linspace(10, 100, 100)

resistance_data = []

for width in Wcu:
    for thickness in Tcu:
        Wcu_m = width * 1e-9
        Tcu_m = thickness * 1e-9

        resistance = rowcu / (Tcu_m * Wcu_m)
        resistance_data.append([width, thickness, resistance])


np.savetxt("copper_resistance_data.csv", resistance_data, delimiter=",", header="Width (nm), Thickness (nm), Resistance (Ohms)", fmt='%f, %f, %.4e', comments='')
```
