```python
df_final_cap = pd.DataFrame(columns=['Copper Capacitance (F)', 'Graphene Capacitance (F)', 'Final Capacitance (F)'])


for index, row in df_cu_cap.iterrows():
    ccu = row[' Capacitance (F)']
    cgr = df_gr_cap.iloc[index]['Cg']

    cesc = 1 / ((1 / cgr) + (1 / ccu))

    df_final_cap.loc[index] = [ccu, cgr, cesc]


df_final_cap.to_csv('/content/CuMlg_capacitance.csv', index=False)
```

```python
df_final_ind = pd.DataFrame(columns=['Copper Inductance (H)', 'Graphene Inductance (H)', 'Final Inductance (H)'])


for index, row in df_cu_ind.iterrows():
    lcu = row[' Inductance (H)']
    lgr = df_gr_ind.iloc[index]['Lg']

    lesc = 1 / ((1 / lgr) + (1 / lcu))

    df_final_ind.loc[index] = [lgr, lcu, lesc]

df_final_ind.to_csv('/content/CuMlg_inductance.csv', index=False)
```

```python
N = 3
df_final_res = pd.DataFrame(columns=['Copper Resistance (Ohms)', 'Graphene Resistance (Ohms)', 'Final Resistance (Ohms)'])

for index, row in df_cu_res.iterrows():
    rcu = row[' Resistance (Ohms)']
    rgr = df_gr_res.iloc[index]['Rgr']
    resc = 1 / ((1 / rcu) + N * (1 / rgr))

    df_final_res.loc[index] = [rcu, rgr, resc]


df_final_res.to_csv('/content/CuMlg_resistance.csv', index=False)
```
