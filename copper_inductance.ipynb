```python
import numpy as np

μ0 = 4 * np.pi * 1e-7
Lcu = np.linspace(50, 200, 100)
Tcu = np.linspace(10, 100, 10)
Wcu = np.linspace(50, 140, 10)

inductance_data = []

for length in Lcu:
    for thickness in Tcu:
        for width in Wcu:
            lc_m = length * 1e-9
            tc_m = thickness * 1e-9
            wc_m = width * 1e-9

            inductance = μ0 / (2 * np.pi) * (np.log(2 * lc_m / (tc_m + wc_m)) + 0.5 + 0.22 * (tc_m + wc_m) / lc_m)
            inductance_data.append([length, thickness, width, inductance])


np.savetxt("copper_inductance_data.csv", inductance_data, delimiter=",", header="Length (nm), Thickness (nm), Width (nm), Inductance (H)", fmt='%f, %f, %f, %.12e', comments='')
```
